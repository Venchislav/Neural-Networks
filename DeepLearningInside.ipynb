```python
import numpy as np
import time
import random
```

```python
x = np.random.random((20, 100))
y = np.random.random((20, 100))
```

```python
 t0 = time.time()
```

```python
for _ in range(1000):
  z = x + y
  z = np.maximum(z, 0.)
print(time.time() - t0)
```

```python
x = [[random.random() for _ in range(20)] for _ in range(100)]
y = [[random.random() for _ in range(20)] for _ in range(100)]
```

```python
def w_sum(a, b):
  assert (len(a) == len(b)) and (len(a[0]) == len(b[0]))

  for i in range(len(a)):
    for j in range(len(a[0])):
      z = a[i][j] + b[i][j]
      z = max(z, 0)
```

```python
t0 = time.time()
for _ in range(1000):
  w_sum(x, y)

print(time.time() - t0)
```

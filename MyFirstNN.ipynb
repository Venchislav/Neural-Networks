# **EXAMPLE OF GRADIENT DESCENT IN SIMPLE NN WITH 1 INPURT AND 1 OUTPUT**

<img src='https://gbhat.com/assets/gifs/gradient_descent.gif' width=400>

```python
weight = 0.0
goal_pred = 0.8
input = 1.1
```

```python
pred = float('inf')
while pred != goal_pred:
  print(f'Weight {weight}')
  # getting prediction
  pred = input * weight
  # getting error
  err = (pred - goal_pred) ** 2
  # delta (err but not squared)
  delta = pred - goal_pred
  # weight delta is just scaled delta as we work with gradient descent and it's simmilar to pendulum
  weight_delta = delta * input
  # changing the weight on weight delta
  # if we predicted 1, and goal is 0.8 delta will be  0.2, weight_delta will be 0.1, weight will change on - 0.1)
  weight -= weight_delta

  print(f'Prediction {pred} | Delta {delta} | Weight Delta {weight_delta} | Error {err}\n{"-"*60}')
```

```python
# this was the simplest possible neural network with gradient descent
```

```python
for _ in range(20):
  pred = input * weight
  err = (pred - goal_pred) ** 2
  delta = pred - goal_pred
  weight_delta = delta * input
  weight -= weight_delta

  print(pred)
```

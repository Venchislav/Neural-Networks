```python
import tensorflow as tf
import cv2
import pandas as pd
import numpy as np
```

```python
train_labels = pd.read_csv('/kaggle/input/game-of-deep-learning-ship-datasets/train/train.csv')
train_labels.head()
```

```python
X = []
y = []
```

```python
for name in train_labels['image']:
    img = cv2.imread(f'/kaggle/input/game-of-deep-learning-ship-datasets/train/images/{name}')
    X.append(np.asarray(img))
    y.append(train_labels.iloc[np.where(train_labels['image'] == name)[0][0], 1])
```

```python
y[:5]
```

```python
X = np.array(X, dtype=np.ndarray)
y = np.array(y)
```

```python
import matplotlib.pyplot as plt

plt.imshow(X[0])
```

```python
len(y)
```

```python
len(X)
```

```python
IM_SIZE = 224

def resize_rescale(img):
    return tf.image.resize(img, (IM_SIZE, IM_SIZE)) / 255.0
```

```python
X = list(map(resize_rescale, X))
```

```python
plt.imshow(X[0])
```

```python
X = np.array(X)
```

```python
for i in range(9):
    plt.subplot(3, 3, i + 1)
    plt.title(y[i])
    plt.axis('off')
    plt.imshow(X[i]);
```

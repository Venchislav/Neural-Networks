```python
import tensorflow as tf
import cv2
import pandas as pd
import numpy as np
```

```python
train_labels = pd.read_csv('/kaggle/input/game-of-deep-learning-ship-datasets/train/train.csv')
train_labels.head()
```

```python
X = []
y = []
```

```python
for name in train_labels['image']:
    img = cv2.imread(f'/kaggle/input/game-of-deep-learning-ship-datasets/train/images/{name}')
    X.append(np.asarray(img))
    y.append(train_labels.iloc[np.where(train_labels['image'] == name)[0][0], 1])
```

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('application_record.csv.zip')
record = pd.read_csv('credit_record.csv')
df.head()
```

```python
df = pd.merge(df, record,on='ID', how='outer')
```

```python
df['OCCUPATION_TYPE'].value_counts()
```

```python
df = df.dropna()
```

```python
df.isnull().sum()
```

```python
df.dtypes
```

```python
df = df.drop('ID',axis='columns')
```

```python
df.shape
```

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import MinMaxScaler
```

```python
encoder = LabelEncoder()
```

```python
obj_vals = df.select_dtypes(include='object')
```

```python
obj_vals.columns
```

```python
for c in obj_vals.columns:
    df[c] = encoder.fit_transform(df[c])
```

```python
df.head()
```

```python
df.dtypes
```

```python
df = df.head(10_000)
```

```python
X = df.drop('STATUS', axis=1)
y = df['STATUS']
```

```python
scaler = MinMaxScaler((0, 1))
X = scaler.fit_transform(X)
```

```python
X.max()
```

```python
from sklearn.utils import shuffle
```

```python
X, y = shuffle(X, y)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
print(X_train.shape, '-' * 10, y_train.shape)
print('=' * 60)
print(X_test.shape, '-' * 10, y_test.shape)
```

```python
inp_shape = X_train.shape[1:]
```

```python
print(inp_shape)
```

```python
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.models import Sequential
```

```python
model = Sequential([
    layers.Dense(units=26, activation='relu', input_shape=inp_shape),
    layers.Dropout(0.2),
    layers.Dense(units=36, activation='relu'),
    layers.Dropout(0.2),
    layers.Dense(units=46, activation='relu'),
    layers.Dropout(0.2),
    layers.Dense(units=64, activation='relu'),
    layers.Dropout(0.2),
    layers.Dense(units=8, activation='softmax'),
])
```

```python
model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)
```

```python
hist = model.fit(
    X_train, y_train,
    validation_split=0.3,
    verbose=2,
    shuffle=True,
    batch_size=512,
    epochs=500
)
```

```python
pred = model.predict(X_test)
```

```python
y_pred = []
for i in range(pred.shape[0]):
    y_pred.append(np.argmax(pred[i]))
```

```python
len(y_pred)
```

```python
y_test[:5]
```

```python
from sklearn.metrics import accuracy_score
```

```python
print(accuracy_score(y_test, y_pred))
```

```python
import numpy as np
```

```python
def activation(x):
  return 0 if x < 0.5 else 1
```

```python
def go(house, rock, attr):
  arr = np.array([house, rock, attr])
  wi_1 = [0.3, 0.3, 0]
  wi_2 = [0.4, -0.5, 1]
  weight1 = np.array([wi_1, wi_2])
  weight2 = np.array([-1, 1])

  sum_hidden = np.dot(weight1, arr)
  print(f"Neuron's sum values: {sum_hidden}")

  out_hidden = np.array([activation(x) for x in sum_hidden])
  print(f"Hidden's layer output: {out_hidden}")

  sum_end = np.dot(weight2, out_hidden)
  y = activation(sum_end)

  return f'Output for this neural network is: {y}'
```

```python
house = 0
rock = 0
attr = 1
```

```python
go(house, rock, attr)
```
